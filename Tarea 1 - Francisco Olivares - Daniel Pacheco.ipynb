```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import sklearn.linear_model as lm
import pylab
import scipy.stats as stats
from sklearn.model_selection import KFold
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline
def parte1(arreglar = False):
    #=====================================================
    #PARTE A
    #=====================================================
    print("===================================")
    print("INICIO DE LA 1 A")
    print("===================================")
    df = pd.read_csv("kc_house_data.csv")
    df.drop(['id','date','zipcode',],axis=1,inplace=True)
    
    #IMPORTANTE PARA LA PREGUNTA F
    if arreglar == True:
        del df["sqft_above"]
        
    display(df.head())
    #=====================================================
    #PARTE B
    #=====================================================
    df.shape
    df.info()
    display(df.describe())
    #=====================================================
    #PARTE C
    #=====================================================
    print("===================================")
    print("INICIO DE LA 1 C")
    print("===================================")
    scaler = StandardScaler()
    df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
    
    plt.figure(figsize = (16, 5))
    plt.hist(df_scaled['price'], bins = 10)
    plt.show()
    
    df_scaled['price'] = np.log(df['price'])
    
    plt.figure(figsize = (16, 5))
    plt.hist(df_scaled['price'], bins = 10)
    plt.show()
    #=====================================================
    #PARTE D
    #=====================================================

    X = df_scaled.iloc[:,1:] #use .ix instead, in older pandas version
    N = X.shape[0]
    X.insert(X.shape[1], 'intercept', np.ones(N))
    y = df_scaled['price']
    #mascara estatica con el 70% de los datos
    mascara = np.zeros(len(X))
    limit = int(len(X)*0.7)
    mascara[:limit] = 1 
    istrain = mascara== 1
    Xtrain = X[istrain]
    ytrain = y[istrain]
    Xtest = X[np.logical_not(istrain)]
    ytest = y[np.logical_not(istrain)]
    linreg = lm.LinearRegression(fit_intercept = False)
    modeloLineal = linreg
    linreg.fit(Xtrain, ytrain)
    
    #========================================================================================
    #PARTE E
    #========================================================================================
    print("===================================")
    print("INICIO DE LA 1 E")
    print("===================================")
    pesos = linreg.coef_ #el vector beta

    yEstimado = linreg.predict(Xtrain)
    yReal = np.array(ytrain)

    #Procedo a estimar la desviaci√≥n estandar
    std = np.sqrt(sum(map(lambda x: x**2, (yReal - yEstimado)))) / (len(yReal) - len(pesos))

    #Procedo a sacar la diagonal de la pseudo matriz de covarianzas
    MatrizXTX = np.dot(np.asarray(Xtrain).T , np.asarray(Xtrain))

    V = np.linalg.inv(MatrizXTX).diagonal()
    valoresPropios = np.linalg.eigvals(MatrizXTX)
    print("Los valores propios son\n", valoresPropios)

    zScore = list()
    for i in range(len(pesos)):
        zScore.append(pesos[i] / (std * np.sqrt( V[i] )))

    a = list(df)[1:]
    a.append("origen")

    respuesta = [("Atributo", a ),("Peso", pesos) , ("Z-Score", zScore), ("Diagonal", V)]
    display(pd.DataFrame.from_items(respuesta))
    #=======================================================================================
    #PARTE F
    #=======================================================================================
    if arreglar == False:
        print("===================================")
        print("INICIO DE LA 1 F")
        print("===================================")    
        sospechosos = ["sqft_living", "sqft_living15" , "sqft_lot", "sqft_above", "sqft_basement"]
        for i in range(len(sospechosos)):
            for j in range(len(sospechosos)):
                if i<j:
                    x = np.asarray(df[sospechosos[i]])
                    y = np.asarray(df[sospechosos[j]])
                    plt.figure(figsize = (16, 5))
                    plt.scatter(x, y)            
                    plt.xlabel(sospechosos[i])
                    plt.ylabel(sospechosos[j])
                    plt.show()
    else:
        print("No se muestran los gr√°ficos ya que se arregl√≥ el problema c:")
    #
    #PARTE G
    #
    print("===================================")
    print("INICIO DE LA 1 G")
    print("===================================")
    yhat_test = linreg.predict(Xtest)
    mse_test = np.mean(np.power(yhat_test - ytest, 2))
    Xm = Xtrain.as_matrix()
    ym = ytrain.as_matrix()
    def cve(k):
        kf = KFold(n_splits=k)
        mse_cv = 0
        for train, val in kf.split(Xm):
            linreg = lm.LinearRegression(fit_intercept = False)
            linreg.fit(Xm[train], ym[train])
            yhat_val = linreg.predict(Xm[val])
            mse_fold = np.mean(np.power(yhat_val - ym[val], 2))
            mse_cv += mse_fold
        mse_cv = mse_cv / k
        return mse_cv
    k_5 = cve(5)
    k_10 = cve(10)
    k_15 = cve(15)
    
    respuesta = [("15 Folds", [k_15] ),("10 Folds", [k_10]) , ("5 Folds", [k_5]), ("Error Prueba", [mse_test])]
    display(pd.DataFrame.from_items(respuesta))
    
    #print("Folds 5: "+str(k_5)+"\nFolds 10: "+str(k_10)+"\nError de entrenamiento: "+str(mse_test) )
    #
    #PARTE H
    #
    print("===================================")
    print("INICIO DE LA 1 H")
    print("===================================")
    yEstimado = np.asarray(modeloLineal.predict(Xtrain))
    yReal = np.asarray(ytrain)

    diferencias = yReal - yEstimado #np.power(yReal - yEstimado, 2)
    diferencias.sort()

    plt.figure(figsize = (16, 5))
    #measurements = np.random.normal(loc = 20, scale = 5, size=100)
    #display(measurements)
    stats.probplot(diferencias, dist="norm", plot=pylab)
    pylab.show()
    
parte1()#parte1(arreglado = True)
```

### d)
<span style="font-size:12pt">
El modelo de regresi√≥n lineal es
$$ y = \beta_0 + \sum_{j=1}^{p} \beta_j x_j  $$
Se puede meter el termino $\beta_0$ dentro del vector $\beta$ si es que al final de cada muestra (matriz $X$) se a√±ade un 1, entonces al momento de resolver el sistema $X \cdot \beta = Y$, dentro del vector $\beta$ la √∫ltima fila ser√° el $\beta_0$.<br>
Lo que hace la linea 4 es exactamente a√±adir una columna de 1's a la matriz $X$.<br>
<br>
Los par√°metros usados en la regresi√≥n lineal<br>
Para la funci√≥n `lm.LinearRegression()`<br>
* fit_intercept = False. Sirve para ver si se va a calcular o no el intercepto, por defaul True<br>
* normalize = Se ignora si fit_intercept es False. Si es True la data X va a ser normalizada<br>
* copy_X = Boleano que dice si las operaciones ser√°n o no inplace (default True)<br>
* n_jobs = los jobs usados para el procesamiento (default True)<br>

<span style="font-size:12pt">
Para el m√©todo `fit()`<br>
* X = La matriz $X$
* y = El vector y de soluciones conocidas para el input $X$
</span>

### e)
<span style="font-size:12pt">
El z-score se calcula como $$\textit{z-score} = \frac{\hat{\beta_j}}{\hat{\sigma} \cdot \sqrt{\upsilon_{jj}}}$$
<span style="font-size:12pt">
Donde $\hat{\beta_j}$ corresponden a los pesos del atributo j, $\hat{\sigma}$ es la desviaci√≥n estandar calculada de la forma $$\hat{\sigma} = \frac{\sqrt{ \sum_{i=1}^{n} (y^{(i)} - \hat{y}^{(i)})^2}} {n - d - 1}$$
<span style="font-size:12pt">
y $\upsilon_{jj}$ corresponde a la diagonal de la pseudo matriz de covarianzas (le faltar√≠a el $\frac{1}{n}$), es decir $$\upsilon_{jj} = (X^{T} \cdot X)^{-1} _{jj}$$
Entonces as√≠ se obtuvo la tabla de arriba.
<span style="font-size:12pt">
Se observa que 3 de los valores del z-score dan NaN, esto ocurre dado que algunos $\upsilon_{jj}$ son negativos. Esto se produce ya que la Matriz $X^{T} \cdot X$ esta al borde de ser singular dado que uno de sus valores propios est√° muy cerca del 0 (-2.29675177e-12), entonces al momento de sacar la inversa se generan esos $\upsilon_{jj}$ negativos.
<span style="font-size:12pt">
Notar que si uno de los valores propios est√° muy cerca del 0 eso significa que alguno o alguno(s) atributos son linealmente dependientes.<br><br>
Los z-score con magnitud mayor a 2 son los atributos significativos, mientras m√°s z-score m√°s significativo.
</span>

# 2 Selecci√≥n de atributos

```python
def parte2():
    df = pd.read_csv("kc_house_data.csv")
    df.drop(['id','date','zipcode','sqft_above'],axis=1,inplace=True)
    scaler = StandardScaler()
    df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
    df_scaled['price'] = np.log(df['price'])
    X = df_scaled.iloc[:,1:] #use .ix instead, in older pandas version
    N = X.shape[0]
    X.insert(X.shape[1], 'intercept', np.ones(N))
    y = df_scaled['price']
    #mascara estatica con el 70% de los datos
    mascara = np.zeros(len(X))
    limit = int(len(X)*0.7)
    mascara[:limit] = 1 
    istrain = mascara== 1
    Xtrain = X[istrain]
    ytrain = y[istrain]
    Xtest = X[np.logical_not(istrain)]
    ytest = y[np.logical_not(istrain)]
    def fss(x, y, names_x, k = 10000):     
        p = x.shape[1]-1 #numero de columnas - 1
        k = min(p, k) #cantidad maxima de atributos
        names_x = np.array(names_x) 
        remaining = [i for i in range(0, p)] #lista de 0 a 16 considerando los 17 primeros
        selected = [p] #parto seleccionando el intercepto
        current_score = best_new_score = 0.0
        ejeX = [i for i in range(1,p+1)]
        erroresEntrenamiento = list()
        erroresPrueba = list()
        while remaining and len(selected)<=k :
            score_candidates = []
            for candidate in remaining:
                model = lm.LinearRegression(fit_intercept=False)
                indexes = selected + [candidate]
                x_train = x[:,indexes]

                x_test = Xtest.as_matrix()[:,indexes]
                y_test = ytest.as_matrix()

                modeloEntrenado = model.fit(x_train, y)
                predictions_train = modeloEntrenado.predict(x_train)
                residuals_train = predictions_train - y

                residuoTest = modeloEntrenado.predict(x_test) - y_test
                #rmse_test = np.sqrt(sum(np.power(residuoTest, 2)) * (1.0/len(residuoTest)))

                mse_candidate = np.mean(np.power(residuals_train, 2))
                mse_test = np.mean(np.power(residuoTest, 2))

                rmse_candidate = np.sqrt(sum(np.power(residuals_train, 2)) * (1.0/len(residuals_train)))
                
                #=========================================================================================
                #SE PROCEDE A SACAR EL Z-SCORE
                #=========================================================================================
                pesos = modeloEntrenado.coef_ #el vector beta

                yEstimado = predictions_train
                yReal = y

                #Procedo a estimar la desviaci√≥n estandar
                std = np.sqrt(sum(map(lambda x: x**2, (yReal - yEstimado)))) / (len(yReal) - len(pesos))

                #Procedo a sacar la diagonal de la pseudo matriz de covarianzas
                MatrizXTX = np.dot(np.asarray(x_train).T , np.asarray(x_train))

                V = np.linalg.inv(MatrizXTX).diagonal()

                zScore = list()
                for i in range(len(pesos)):
                    zScore.append(abs(pesos[i] / (std * np.sqrt( V[i] ))))
                zScore = sum(zScore)
                #==========================================================================================
                #FIN DEL CALCULO DEL Z-SCORE
                #==========================================================================================
                
                score_candidates.append((zScore, mse_candidate, candidate, mse_test))

            score_candidates.sort()
            #score_candidates[:] = score_candidates[::-1]        
            _, best_new_score, best_candidate, errorDePrueba = score_candidates.pop()

            erroresEntrenamiento.append(best_new_score)
            erroresPrueba.append(errorDePrueba) 

            remaining.remove(best_candidate)
            selected.append(best_candidate)
            #print ("selected = %s ..."%names_x[best_candidate])
            #print ("totalvars=%d, rmse = %f"%(len(indexes),best_new_score))

        plt.figure(figsize = (16, 5))

        plt.plot(ejeX, erroresEntrenamiento, ".-")
        plt.plot(ejeX, erroresPrueba, ".-")

        plt.legend(["Entrenamiento", "Prueba"])
        plt.xlabel("Numero de atributos")
        plt.ylabel("Errores")

        plt.show()

        return selected
    names_regressors = Xtrain.columns[:-1] #without intercept
    print(fss(Xtrain.as_matrix(),ytrain.as_matrix(),names_regressors))
parte2()
```

El enfoque utilizado es similar al de Backward stepwise selection.

```python
def parte4():
    df = pd.read_csv("kc_house_data.csv")
    df.drop(['id','date','zipcode',],axis=1,inplace=True)
    scaler = StandardScaler()
    df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
    df_scaled['price'] = np.log(df['price'])
    X = df_scaled.iloc[:,1:] #use .ix instead, in older pandas version
    N = X.shape[0]
    X.insert(X.shape[1], 'intercept', np.ones(N))
    y = df_scaled['price']
    #mascara estatica con el 70% de los datos
    mascara = np.zeros(len(X))
    limit = int(len(X)*0.7)
    mascara[:limit] = 1 
    istrain = mascara== 1
    Xtrain = X[istrain]
    ytrain = y[istrain]
    Xtest = X[np.logical_not(istrain)]
    ytest = y[np.logical_not(istrain)] 
    
    df_A = df_scaled.sample(1000,random_state=11)
    frames = []
    valor = df_scaled.price
    length = 0.3
    for z in np.arange(int(np.min(valor)),int(np.max(valor))+1,length):
        aux = df_scaled[(df_scaled.price >= z) & (df_scaled.price < z+length)].head(100)
        frames.append(aux)
    df_B = pd.concat(frames).sample(1000,random_state=11)
    
    
    X_A = df_A.iloc[:,1:].values
    y_A = df_A.price
    X_B = df_B.iloc[:,1:].values
    y_B = df_B.price
    from sklearn.model_selection import train_test_split
    Xtrain_A,Xval_A,ytrain_A,yval_A = train_test_split(X_A, y_A, test_size=0.3, random_state=42)
    Xtrain_B,Xval_B,ytrain_B,yval_B = train_test_split(X_B, y_B, test_size=0.3, random_state=42)
    
    
    model = lm.LinearRegression(fit_intercept=True)
    model.fit(Xtrain_A, ytrain_A)
    predictionsVal_AA = model.predict(Xval_A)
    residualsVal_AA = predictionsVal_AA - yval_A
    mse_AA = np.mean(np.power(residualsVal_AA, 2))
    print("Error train A en validaci√≥n A:", mse_AA)

    predictionsVal_AB = model.predict(Xval_B)
    residualsVal_AB = predictionsVal_AB - yval_B
    mse_AB = np.mean(np.power(residualsVal_AB, 2))
    print("Error train A en validacion B:", mse_AB)

    model = lm.LinearRegression(fit_intercept=True)
    model.fit(Xtrain_B, ytrain_B)
    predictionsVal_BA = model.predict(Xval_A)
    residualsVal_BA = predictionsVal_BA - yval_A
    mse_BA = np.mean(np.power(residualsVal_BA, 2))
    print("Error train B en validaci√≥n A:", mse_BA)

    predictionsVal_BB = model.predict(Xval_B)
    residualsVal_BB = predictionsVal_BB - yval_B
    mse_BB = np.mean(np.power(residualsVal_BB, 2))
    print("Error train B en validaci√≥n B:", mse_BB)
    
    #print("Varianza set A", np.var(y_A))
    #print("Varianza set B", np.var(y_B))
    
parte4()
```

```python
from sklearn.metrics import accuracy_score
def parte5():
    #================================================
    #PARTE A
    #================================================
    headers = ['age','sex','chest_pain','blood_p','serum','blood_s','electro','max_heart', 'angina','oldpeak','slope','vessel','thal','normal']
    df = pd.read_csv("heart.dat", header=None, names=headers, sep=' ')
    
    #display(df.head())
    #scaler = StandardScaler()
    #df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

    # Cheackear como se distribuye el atributo
    #plt.figure(figsize = (16, 5))
    #plt.hist(df_scaled['max_heart'], bins = 10)
    #plt.show()
    
    X = df.drop(["max_heart","normal"], axis = 1, inplace = False) #Gran matriz X que le quite el atributo del precio
    #use .ix instead, in older pandas version
    N = X.shape[0]#Cantidad de filas
    #X.insert(X.shape[1], 'intercept', np.ones(N))#Pega una columna de 1's
    y = df["max_heart"]
    
    mascara = np.zeros(len(X))
    limit = int(len(X)*0.7)
    mascara[:limit] = 1
    istrain = mascara== 1
    Xtrain = X[istrain]
    ytrain = y[istrain]
    Xtest = X[np.logical_not(istrain)]
    ytest = y[np.logical_not(istrain)]
    #===================================================
    #PARTE B
    #===================================================
    linreg = lm.LinearRegression(fit_intercept = True, normalize = True)
    linreg.fit(Xtrain, ytrain)    

    df_entrenamiento = df[istrain]
    df_prueba = df[np.logical_not(istrain)]
    a = df_entrenamiento.index[df_entrenamiento["normal"] == 1]
    b = df_entrenamiento.index[df_entrenamiento["normal"] == 2]
    sanos = df_entrenamiento["max_heart"].iloc[a]
    enfermos = df_entrenamiento["max_heart"].iloc[b]
    
    umbral = (np.mean(np.asarray(sanos)) + np.mean(np.asarray(enfermos)))/2.0
    
    def clasificar(x, u):
        if x>=u: 
            return 1
        else:
            return 2
    predicciones = [clasificar(i, umbral) for i in list(linreg.predict(Xtrain))]    
    print ("Score entrenamiento: ",(accuracy_score(list(df_entrenamiento["normal"]),predicciones)))
    
    prediccionesPrueba = [clasificar(i, umbral) for i in list(linreg.predict(Xtest))]   
    print ("Score prueba: ",(accuracy_score(list(df_prueba["normal"]),prediccionesPrueba)))
    
parte5()
```
